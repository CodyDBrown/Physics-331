```python
import numpy as np
import matplotlib.pyplot as plt

L = 1e-10 # Width of the box (m)
A = np.sqrt(2/L) # Normalization constant for the wavefunction.
```

```python
k_list = np.linspace(0, 2*np.pi/L, 11) # List of 'k' values I want to try.
x = np.linspace(0, L, 1000)
plt.figure(figsize = (12,10))
for k in k_list:
    plt.plot(x, A*np.sin(k*x), label = 'k = %.2e'%k)    
    plt.legend()
plt.xlabel('Possition (m)', size = 14)
plt.ylabel('Amplitude of the wavefunction $1/\sqrt{m}$', size = 14)
plt.title("Wavefunction for particle in a box, different wavenumbers 'k'", size = 14)
plt.show()
```

Looking at the graph it looks like k = 0 works, but that gives me the trivial solutions, k = 3.14e+10 works, and so does k = 6.28e+10. Looking at these in terms of multiples of $\pi/L$ we see

```python
k_list[5]/(np.pi/L), k_list[10]/(np.pi/L)
```

```python
x = np.linspace(0, L, 2000)
plt.figure(figsize = (12,10))
k_list = np.linspace(2*np.pi/L, 4*np.pi/L, 11)
for k in k_list:
    plt.plot(x, A*np.sin(k*x), label = 'k = %.2e'%k)    
    plt.legend()
    plt.grid()
plt.xlabel('Possition (m)', size = 14)
plt.ylabel('Amplitude of the wavefunction $1/\sqrt{m}$', size = 14)
plt.title("Wavefunction for particle in a box, different wavenumbers 'k'", size = 14)
plt.show()
```

Now I want to plot the end point of our wavefunction $\psi(L)$ as a function of wave number 'k'

```python
x = np.linspace(0, L, 2000)
plt.figure(figsize = (10,10))
k = np.linspace(0, 10*np.pi/L, 10000)

plt.plot(k, A*np.sin(k*L), label = '$\psi(L)$') 
plt.plot(np.arange(10)*np.pi/L, np.zeros(10), 'x', color = 'r', label = '$n*\pi / L$')

plt.title('End point of the wavefunction, as a function of "k"', size = 14)
plt.ylabel('$\psi(L)(k)$', size = 14)
plt.xlabel("Wave number 'k' (1/m)", size = 14)
plt.grid()
plt.legend()
plt.show()    
```

It looks like $\psi(L)$ crosses zero when $k = n * \pi/L$. We could use this with a function like bisect.py to solve for what 'k' values satisfy the second boundry condition.
